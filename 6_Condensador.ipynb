```
#Este programa es más raro, y usa otras cosas como separar el plano xy en una rejilla muy fina y asignarle un valor a cada punto
#No es nada que vayáis a hacer en mn1, pero si que os pedirán algo parecido (voluntario) en lab electro
#No es el mejor código para hacer esto, hay gente que los hizo mucho mejores, pero con suerte al menos os servirá de inspiración o para
#cambiarlo un poco y pegarlo en vuestro informe.
#Básicamente la idea es obtener el potencial en el plano de un conensador planoparalelo, poniendo las placas del consedador como suma
#de una cantidad enorme de cargas puntuales muy juntas. Como nos encontramos en 2D, la fuerza electrica decae con 1/r en vez de 1/r^2,
#así que el potencial va con log(r). Ponemos las placas de manera que esten en x=-3 y en x=+3, separadas 6 unidades y, empezando por
#y=-3, calculamos el potencial de un par de cargas en (-3,-3) y (3,-3), le sumamos el de unas cargas en (-3,-2.99999) y (3,-2.99999),
#y así vamos sumando hasta las de (-3,3) y (3,3)

import numpy as np
import matplotlib.pyplot as plt


xx = np.linspace(-14,14)
yy = np.linspace(-10,10)

X, Y = np.meshgrid(xx, yy)

a = -3
Z = (np.log(np.sqrt((X+3)**2 + (Y-a)**2)) - np.log(np.sqrt((X-3)**2 + (Y-a)**2)))

while a<3:
  a=a+0.00001  
  Z = Z+(np.log(np.sqrt((X+3)**2 + (Y-a)**2)) - np.log(np.sqrt((X-3)**2 + (Y-a)**2)))

plt.contourf(X, Y, Z, 30)

fig = plt.subplot()

fig.set_title('Línies equipotencials d\'un condensador de plaques planoparal·leles')

plt.savefig('Simulació Cond.png')

plt.show()
```
